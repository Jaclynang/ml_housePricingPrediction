```python
# drop column ID since it's unrelated
hp = hp.drop(columns = ['Id'])
hp.shape
```

```python
# drop the columns with too many null
for col_name, col_data in hp.items():
    if (hp[col_name].isnull().sum()) > (0.1 * hp.shape[0]):
        hp = hp.drop(
            columns = [col_name],
            axis = 1,
        )

hp.shape
```

```python
# get the sum of missing value of each feature
nana = hp.isnull().sum()

# print only the feature with missing values since there are too many features
colnana = nana[nana > 0]

if not colnana.empty:
    print("Columns with missing values:")
    print(colnana)
else:
    print("No missing values found.")
```

```python
# impute the missing values , mode for categorical, median for numerical data
for col_name, col_data in hp.items():
    if hp[col_name].dtype == 'object':
        hp[col_name] = hp[col_name].fillna(hp[col_name].mode()[0])
    else:
        hp[col_name] = hp[col_name].fillna(hp[col_name].median())
```

```python
# check if there is any missing value left
nana2 = hp.isnull().sum()

colnana2 = nana2[nana2 > 0]

if not colnana2.empty:
    print("Columns with missing values:")
    print(colnana2)
else:
    print("No missing values found.")
```

```python
hp.to_csv('/content/housingprice2.csv', index=False)
```

```python
# use LabelEncoder to label encode the data, one-hot encoder will expand the columns to a large number
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col_name, col_data in hp.items():
    if hp[col_name].dtype == 'object':
        hp[col_name] = le.fit_transform(hp[col_name])

hp.head()
```

```python
# visualise the heatmap of the correlation matrix

plt.figure(figsize = [20, 20])
sns.set(font_scale = 0.25)
sns.heatmap(
    hp.corr(),
    annot = True,
    cmap = 'coolwarm'
)

plt.show()
```

```python
# print the correlation matrix in word as well
corrie = hp.corr()
corrie
```

```python
# get the highly correlated features with SalePrice

threshold = 0.1

# fifi is the features that has higher than 0.1 correlation with SalePrice
fifi = corrie[abs(corrie['SalePrice']) > threshold].index

# ignore SalePrice itself
fifi = fifi.drop('SalePrice', errors='ignore')

# Print the features and the amount of the features
print("Highly correlated features with 'SalesPrice':")
print(fifi)
print(len(fifi))
```

```python
# declare X and Y dataset, fifi is the selected features
X = hp[fifi]
Y = hp['SalePrice']
```

```python
# feature scaling using min max
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X2 = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
X2
```

```python
# split the dataset into train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X2, Y, test_size = 0.2, random_state = 42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

# Extra-Trees Regressor

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

etr = ExtraTreesRegressor()
etr.fit(X_train, Y_train)
etr_pred = etr.predict(X_test)

mae = mean_absolute_error(Y_test, etr_pred)
mse = mean_squared_error(Y_test, etr_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, etr_pred)

print("Mean absolute error: %.4f" %mae)
print("Mean Squared Error (MSE): %.4f" %mse)
print("Root Mean Squared Error (RMSE): %.4f" %rmse)
print("R2 Score: %.4f" %r2)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators' : [25, 50, 100, 150, 200],
    'max_depth' : [10, 12, 14, 15, 16, 18, 20],
    'min_samples_split' : [2, 5, 10],
    'min_samples_leaf' : [1, 2, 4],
}
grid_search = GridSearchCV(etr, param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(X_train, Y_train)

print(grid_search.best_params_, grid_search.best_score_)
```

```python
etr2 = ExtraTreesRegressor(max_depth = 18, min_samples_leaf = 1, min_samples_split = 5, n_estimators = 25)
etr2.fit(X_train, Y_train)
etr_pred2 = etr2.predict(X_test)

mae = mean_absolute_error(Y_test, etr_pred2)
mse = mean_squared_error(Y_test, etr_pred2)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, etr_pred2)

print("Mean absolute error: %.4f" %mae)
print("Mean Squared Error (MSE): %.4f" %mse)
print("Root Mean Squared Error (RMSE): %.4f" %rmse)
print("R2 Score: %.4f" %r2)
```

# **Random Forest Regressor**

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()

rfr.fit(X_train, Y_train)
rfr_pred = rfr.predict(X_test)

mae = mean_absolute_error(Y_test, rfr_pred)
mse = mean_squared_error(Y_test, rfr_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, rfr_pred)

print("Mean absolute error: %.4f" %mae)
print("Mean Squared Error (MSE): %.4f" %mse)
print("Root Mean Squared Error (RMSE): %.4f" %rmse)
print("R2 Score: %.4f" %r2)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 200,250,300],
    'max_depth': [None, 1, 2, 3, 4],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf' : [1, 2, 4],
}

# Create the GridSearchCV object
grid_search = GridSearchCV(rfr, param_grid,cv=5, scoring='r2',n_jobs=-1)

# Fit the grid search to the data
grid_search.fit(X_train, Y_train)
print(grid_search.best_params_, grid_search.best_score_)
```

```python
rfr2 = RandomForestRegressor(n_estimators = 200, max_depth = None, min_samples_leaf=1, min_samples_split=5)
rfr2.fit(X_train, Y_train)
rfr_pred2 = rfr2.predict(X_test)

mae = mean_absolute_error(Y_test, rfr_pred2)
mse = mean_squared_error(Y_test, rfr_pred2)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, rfr_pred2)

print("Mean absolute error: %.4f" %mae)
print("Mean Squared Error (MSE): %.4f" %mse)
print("Root Mean Squared Error (RMSE): %.4f" %rmse)
print("R2 Score: %.4f" %r2)
```
